```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("kanser.csv")
data.head()
```

```python
data.drop(["id","Unnamed: 32"],axis=1,inplace=True)
```

```python
M = data[data.diagnosis == "M"]
B = data[data.diagnosis == "B"]
```

```python
plt.scatter(M.radius_mean, M.texture_mean, color="red", label = "KÃ¶tÃ¼")
plt.scatter(B.radius_mean, B.texture_mean, color="green", label = "Ä°yi",alpha=0.4) #alpha =  saydamlÄ±Ä±k; estetik amaÃ§lÄ±
plt.xlabel("radius mean")
plt.ylabel("texture mean")
plt.legend()
plt.show()
```

```python
data.diagnosis=[1 if each=="M" else 0 for each in data.diagnosis]
```

```python
x=data.iloc[:,1:] #database'deki M ve B'ler
y=data.diagnosis.values # Bizim M ve B'lerden oluÅŸturduÄŸumuz 1 ve 0'lar
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2, random_state=23)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
xtrain1 = sc.fit_transform(xtrain)
xtest1 = sc.transform(xtest)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(xtrain1,ytrain)
```

```python
yhead = lr.predict(xtest1)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest,yhead)
```

```python
import seaborn as sns
f,ax = plt.subplots(figsize = (5,5))
sns.heatmap(cm,annot=True, linewidths=0.5, linecolor="red", fmt = ".0f",ax=ax)
```
